For using the following code example:
Basically you can just change the 'overpass_query' part to use. Don't forget to fill in the longitude & latitude.

```python
import requests
import json

overpass_url = "http://overpass-api.de/api/interpreter"
# overpass_query = """
# [out:json];
# (node["name"="LinkNYC"](40.499479, -74.288986, 40.932319, -73.709685);
#  way["name"="LinkNYC"](40.499479, -74.288986, 40.932319, -73.709685);
#  rel["name"="LinkNYC"](40.499479, -74.288986, 40.932319, -73.709685);
# );
# out center;
# """
overpass_query = """
[out:json];
node["amenity"="restaurant"](40.499479, -74.288986, 40.932319, -73.709685);
out;
way["amenity"="restaurant"](40.499479, -74.288986, 40.932319, -73.709685);
out center;
relation["amenity"="restaurant"](40.499479, -74.288986, 40.932319, -73.709685);
out center;
"""
response = requests.get(overpass_url, 
                        params={'data': overpass_query})
data = response.json()
```
